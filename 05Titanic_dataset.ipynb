### ML implementation on Taitanic Datasets

```python
import pandas as pd 
import numpy as np 
```

#### Read Data sets

```python
df=pd.read_csv('titanic dataset.csv')
df.head()
```

#### Data Cleaning

```python
df.isna().sum()
```

```python
df.drop(columns=['Cabin'],axis=1,inplace=True)
```

```python
df.head(4)
```

```python
df.dtypes
```

```python
df.describe(include='all')
```

```python
#1st methood 
df.replace({'Age':29,
           "Embarked":"s"
           })
```

```python
#2nd method
df.Embarked.replace(np.NaN,"s",inplace=True)
df.Age.replace(np.NaN,29,inplace=True)
```

```python
df.isna().sum()
```

```python
#3rd method
df.Age.fillna(30,inplace=True)
df.Embarked.fillna('S',inplace=True)
```

```python
df.isna().sum()
```

```python
for i in df.columns:
    print(f"{i}:\n{df[i].unique()}\n")
```

```python
df.head(3)
```

```python
#Drop Unnecassary column
df.drop(columns=['PassengerId','Name','Ticket'],index=1,inplace=True)
```

```python
df.head()
```

#### Label encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
for i in df.columns:
    print(f"{i}:\n{df[i].unique()}\n")
```

```python
import math
df.Fare=df.Fare.round().astype(int)
df.Age=df.Age.round().astype(int)
```

```python
df.head()
```

#### Ip/Op creation

```python
x=df.drop(['Survived'],axis=1)
y=df.Survived
```

#### ML implementation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=SVC(gamma='auto',kernel='rbf')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

#### Accuracy Prediction

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
pred = model.predict(x_test)
y_test_pred=accuracy_score(pred,y_test)
print(y_test_pred)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(pred,y_test)
sns.heatmap(cm,annot=True,cmap='Blues')
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```
